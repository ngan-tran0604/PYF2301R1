```python
# Chủ đề: "Quản lý nhân sự. (thông tin nhân sự: họ tên, quê quán, mã phòng ban, lương hiện tại, ngày vào công ty)
#Các chức năng chính: Thêm, tìm kiếm, chỉnh sửa và xóa thông tin"
```

```python
class Person:
    def __init__(self, name, place, idnumber, salary, start_date):
        self.name = name
        self.place = place
        self.idnumber = idnumber
        self.salary = salary
        self.start_date = start_date
        

#Human Resources System

from tkinter import *
from tkinter.messagebox import showinfo
from tkinter import ttk
import datetime as dt
from tkcalendar import DateEntry

window = Tk()
window.title("Human Resources System")
window.geometry("1050x400")
s = ttk.Style()


# Label
intro_lbl = Label(window, text="Human Resources System", fg='blue')
my_salary_lbl = Label(window, text="Salary(VND)")
my_idnumber_lbl = Label(window, text="ID number")
my_name_lbl = Label(window, text="Employee's name")
my_place_lbl = Label(window, text="Place of birth")
my_date_lbl = Label(window, text="Start date")
search_lbl = Label(window, text="Search")

# Table
tree = ttk.Treeview(window, column=("c1", "c2", "c3", "c4", "c5"), show='headings', height=5)

tree.column("# 1", anchor=CENTER)
tree.heading("# 1", text="ID")
tree.column("# 2", anchor=CENTER)
tree.heading("# 2", text="Name")
tree.column("# 3", anchor=CENTER)
tree.heading("# 3", text="Salary")
tree.column("# 4", anchor=CENTER)
tree.heading("# 4", text="Place")
tree.column("# 5", anchor=CENTER)
tree.heading("# 5", text="Date")


# Entry
salary = Entry(window)
idnumber = Entry(window)
name = Entry(window)
place = Entry(window)
# start_date = Entry(window)
start_date = DateEntry(window,selectmode='day')
search_name = Entry(window)

# list_person được sử dụng để lưu trữ thông tin các nhân viên
list_person = []

def add_person():
    # lấy thông tin từ các entry
    salary_value = int(salary.get())
    idnumber_value = int(idnumber.get())
    name_value = name.get()
    place_value = place.get()
    start_date_value = start_date.get()
    
    # khởi tạo đối tượng
    person = Person(name_value, place_value, idnumber_value, salary_value, start_date_value)
    
    # thêm đối tượng vào danh sách
    list_person.append(person)
    
    tree.insert('', 'end', text="1", values=(person.idnumber, person.name, person.salary, person.place, person.start_date))
    
    # Tinh tong salary
    sum = 0
    for person in list_person:
        sum += person.salary
    total_salary_lbl = Label(window,text = f"Total salary per month = {sum}")
    total_salary_lbl.grid(row=11, column=0)
    
    #Tinh tong nhan vien
    total_employee_lbl = Label(window,text = f"Total employees = {len(list_person)}")
    total_employee_lbl.grid(row=12, column=0)
    
    showinfo("Human Resources System", "Person added successfully!")
    

def search_person():
    search_name_value = search_name.get()
    for person in list_person:
        if search_name_value in person.name:
            showinfo("Search result", f"Employee's name: {person.name} - Place of birth: {person.place} - ID number: {person.idnumber} - Salary: {person.salary} - Date: {person.start_date}")
            break
            
def update_person():
    salary_value = int(salary.get())
    idnumber_value = int(idnumber.get())
    name_value = name.get()
    place_value = place.get()
    start_date_value = start_date.get()
    
    for person in list_person:
        if name_value == person.name:
            person.place = place_value
            person.idnumber = idnumber_value
            person.salary = salary_value
            person.start_date = start_date_value
            showinfo("Update result", f"Update information for person {person.name} successfully!")
    clear_all()
    
    for person in list_person:
        tree.insert('', 'end', text="1", values=(person.idnumber, person.name, person.salary, person.place, person.start_date))
        
    
    # Tính tổng salary
    sum = 0
    for person in list_person:
        sum += person.salary
    total_salary_lbl = Label(window,text = f"Total salary per month = {sum}")
    total_salary_lbl.grid(row=11, column=0)
   
            
def clear_all():
    for item in tree.get_children():
        tree.delete(item)

def delete_person():
    search_name_value = search_name.get()
    
    for i in range(len(list_person)):
        if search_name_value == list_person[i].name:
            del list_person[i]
            showinfo("Action result", "Delete person successfully!")
            break
    clear_all()
    
    for person in list_person:
        tree.insert('', 'end', text="1", values=(person.idnumber, person.name, person.salary, person.place, person.start_date))
        
    
    # Tinh tong salary
    sum = 0
    for person in list_person:
        sum += person.salary
    total_salary_lbl = Label(window,text = f"Total salary per month= {sum}")
    total_salary_lbl.grid(row=11, column=0)
    
    #Tinh tong nhan vien
    total_employee_lbl = Label(window,text = f"Total employees = {len(list_person)}")
    total_employee_lbl.grid(row=12, column=0)
    
#Hiển thị ngày tháng
date = dt.datetime.now()
label = Label(window, text=f"{date:%A, %B %d, %Y}", font="Calibri, 12")
label.grid(row=1, column=0, columnspan=5) 


# Button
add_btn = Button(window, text="Add", command=add_person, fg='red')
search_btn = Button(window, text="Search", command=search_person, fg='red')
update_btn = Button(window, text="Update", command=update_person, fg='red')
delete_btn = Button(window, text="Delete", command=delete_person, fg='red')

intro_lbl.grid(row=0, column=0, columnspan=5)
my_name_lbl.grid(row=2,column=0)
name.grid(row=2,column=1)
my_place_lbl.grid(row = 3, column = 0)
place.grid(row = 3, column = 1)
my_idnumber_lbl.grid(row = 4, column = 0)
idnumber.grid(row=4, column=1)
my_salary_lbl.grid(row=5, column=0)
salary.grid(row=5, column=1)
my_date_lbl.grid(row=6, column=0)
start_date.grid(row=6, column=1)
add_btn.grid(row=7, column=1)
update_btn.grid(row=6, column=2)
search_lbl.grid(row=2, column=2)
search_name.grid(row=3, column=2)
search_btn.grid(row=4, column=2)
delete_btn.grid(row=5, column=2)
tree.grid(row = 10, column = 0, columnspan=4)


window.mainloop()
```
