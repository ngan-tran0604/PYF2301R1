## Styling trực tiếp

```python
from tkinter import *
from tkinter import font

#Setup window:
window = Tk()
window.title("Styling")
window.geometry('500x600')
window.configure = '#FAFAFA'

#Define font:
fontTitle = font.Font(family="Stencil Std", size= 44, weight= font.BOLD)
fontText = font.Font(family="Terminal", size= 30)

print(font.families)

#Styling label:
'''Hiển thị bằng pack(), place(), grid()'''

'''width (qui định số ký tự), padx, pady(khoảng cách từ label)'''

''''''
labelTitle = Label(window, 
                   text='Hello',
                   font=fontTitle, 
                   bg='red', 
                   fg='yellow', 
                   width=10, 
                   padx=10, 
                   pady=10
                  )
labelTitle.pack()

labelText = Label(window, 
                  text='Python',
                  font=fontText, 
                  bg='red', fg='yellow', 
                  width=10, 
                  padx=10, 
                  pady=10
                 )
labelText.pack()

#Styling Button:
buttonLogin = Button(window, 
                     text='Log in', 
                     font=fontText, 
                     padx=30, 
                     pady=20, 
                     bg='yellow', 
                     fg='red',
                     activebackground = 'green',
                     activeforeground = 'white'
                    
                    )
buttonLogin.pack()


#Styling Entry:
entryUsername = Entry(window, 
                     text='Log in', 
                     font=fontText,  
                     bg='#FAFAFA'
                     )
'''Chiều cao của entry sẽ phụ thuộc vào kích thước chữ.'''
entryUsername.pack()
entryUsername.focus()

#run
window.mainloop()
```

### Styling trạng thái của button

```python
import tkinter as tk
from tkinter import ttk

#Setup window:
window = tk.Tk()
window.title("Styling Button")
window.geometry('500x600')
window.configure = '#FAFAFA'

#Styling button
style = ttk.Style()
style.configure('TButton', background = 'blue', foreground = 'blue', 
                font=('Terminal', 16))
style.map('TButton',
            foreground=[('pressed', 'red'),('active', 'yellow'), ('disabled', 'gray')],
            background=[('pressed', 'green'),('active', 'yellow')]
          
         )

buttonLogin = ttk.Button(window, text='Log in', style = 'TButton', state = 'active')
buttonLogin.pack()


window.mainloop()
```

## Sử dụng customtkinter để styling

Với CustomTkinter, bạn có thể tạo giao diện người dùng hiện đại trong python bằng tkinter. CustomTkinter là một tiện ích mở rộng của tkinter cung cấp các phần tử giao diện người dùng bổ sung như CTkButton, có thể được sử dụng như một tkinter.Button thông thường, nhưng có thể được tùy chỉnh bằng đường viền và các cạnh tròn.

CustomTkinter cũng hỗ trợ chủ đề sáng và tối, chủ đề này có thể được đặt thủ công hoặc được kiểm soát bởi chế độ giao diện hệ thống (chỉ macOS).

# Demo Form Login

```python
from tkinter import *

#Set up window:
root = Tk()
root.title("Facebook Interface")
root.geometry('600x400')
root.resizable(0, 0)

#Define font:
fontTitle = font.Font(family="Arial", size= 18, weight = font.BOLD)
fontField = font.Font(family="Arial", size= 16)
fontText = font.Font(family="Arial", size= 10)
fontButton = font.Font(family="Arial", size= 12, weight = font.BOLD)

# Tạo Label chữ "Login To Facebook":
name1 = Label(root, text = "Login To Facebook", font = fontTitle, foreground = '#0A89FF').pack(pady = 4)

# Tạo 2 dòng điền username, password bằng Entry, password nhập hiển thị bằng dấu "*" và điều chỉnh khoảng cách.
username_lbl = Label(root, text = "Username", font = fontText).pack(anchor=tk.W, padx = 140)
userbane_entry = Entry(root, font = fontField, width = 30).pack(anchor=tk.W, padx = 140, pady = 8)

password_lbl = Label(root, text = "Password", font = fontText).pack(anchor=tk.W, padx = 140)
password = Entry(root, font = fontField, show = "*", width = 30).pack(anchor=tk.W, padx = 140, pady = 8)

# tạo nút "Login" và chọn màu sắc , kích thước nút:
login_btn = Button(root, text = "Login",  
                   font = fontButton, 
                   bg = "#0A89FF", fg = "#FAFAFA", 
                   padx = 200, pady = 2,
                   activebackground = '#007DF1',
                   activeforeground = 'white'
                  
                  ).pack(anchor=tk.W, padx = 140, pady = 10)

#Tạo dòng chữ"Forgot password?" , chọn màu chữ xanh da trời và điều chỉnh kích thước:
name2 = Label(root, text = "__Forgot password ?__", fg = "#0A89FF").pack()

# tạo dòng kẻ dưới chân nút"Forgot password?":
name3 = Label(root, text = "_____________________________________________________________", fg = "#C1C1C1").pack(pady = (10,0))

#tạo nút tạo một tài khoản mới(Create A New Account) và điều chỉnh kích thước , màu sắc:
name4 = Button(root, text = "Create A New Account", 
               font = fontButton, 
               fg = "white", 
               bg = "#22A433",
               padx = 200, pady = 2,
               activebackground = '#1C9E2D',
               activeforeground = 'white'
              
              ).pack(anchor=tk.W, padx = 140, pady = 10)


#run
root.mainloop()
```
