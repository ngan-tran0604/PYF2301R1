## Câu lệnh if..else

# Đề giống trên

```python
mark = float(input("Nhập điểm cuối kỳ: "))
if mark >= 8.5:
    print("Bạn được tuyển thẳng")
else:
    print("Bạn chưa đủ điểm")
```

### Thực hành:
Viết chương trình nhập vào một số nguyên và cho biết số nguyên đó là số chẵn hay số lẻ.

```python
number = int(input("Nhập số nguyên: "))
if number % 2 == 0:
    print("Số nguyên này là số chẵn")
else:
    print("Số nguyên này là số lẻ")
```

### Thực hành:
Viết chương trình nhập vào một chuỗi và cho biết trong chuỗi đó có chứa cụm từ "python" hay không.

```python
chuoi = input("Nhập vào một chuỗi: ")
if chuoi.count("python") > 0:
    print("có chứa cụm từ python")
else:
    print("không chứa cụm python")
```

## Câu lệnh if lồng nhau

```python
# Nhập vào tổng số giờ làm và level của nhân viên. Nhân viên sẽ được khen thưởng nếu như tổng số giờ làm > 800 giờ. 

# Mức độ khen thưởng như sau:

# - Nếu là nhân viên level 1 thì thưởng 1000

# - Nếu là nhân viên level 2 thì thưởng 2000
```

```python
total_hour = int(input("Nhập tổng số giờ làm: "))
level = int(input("Nhập level của nhân viên: "))
if total_hour > 800:
    if level == 1:
        print("Thưởng 1000")
    else:
        print("Thưởng 2000")
else:
    print("Không thưởng")
```

## Câu lệnh if…elif…else

```python
# Viết chương trình nhập vào số ngày trong tuần (2, 3, 4, 5, 6, 7, 8) và in ra từ tiếng Anh tương ứng với ngày đó.
week_day = int(input("Nhập vào số ngày trong tuần: "))
if week_day == 2:
    print("Monday")
elif week_day == 3:
    print("Tuesday")
elif week_day == 4:
    print("Web")
elif week_day == 5:
    print("Thus")
elif week_day == 6:
    print("Fri")
elif week_day == 7:
    print("Sat")
else:
    print("Sun")
```

## Toán tử ba ngôi

```python
#Nhập vào 1 số n. Cho biết n là số chẵn hay lẻ
n = int(input("Please input n: "))
if n % 2 == 0: print(f"{n} is even number")
else: print(f"{n} is odd number")
```

```python
#Cách khác:
n = int(input("Please input n: "))
if not n % 2 : print(f"{n} is even number")
else: print(f"{n} is odd number")
```

```python
#Cách ngắn hơn của Toán tử ba ngôi:
n = int(input("Please input n: "))
print(f"{n} is even number") if not n % 2 else print(f"{n} is odd number")
```

```python
#Hoặc:
n = int(input("Please input n: "))
print(f"{n} is even number") if n % 2 == 0 else print(f"{n} is odd number")
```
