### Bài 7: Một công ty vận tải có cách tính chi phí giao hàng như sau:

### Bài 10:
Tại một cửa hàng nọ có menu như sau:

1. Gà rán giá 35k
2. Hamburger giá 50k
3. Pesi giá 10k

Nhập số lượng món hàng cần mua

Tính và in ra tổng giá tiền

Nếu giá trị đơn hàng trên 2 triệu, giảm 10% cho khách, trên 4 triệu, giảm 15%

Tính và in ra tổng giá tiền được giảm

Tính và in ra tổng giá tiền khách cần trả

In ra dòng chữ "Cảm ơn quý khách đã mua hàng"

````python
"""Ví dụ:
Nhập số lượng gà rán khách mua:100
Nhập số lượng hamburger khách mua:100
Nhập số lượng pesi khách mua:15
Tổng giá tiền đơn hàng là: 8.650.000
Tổng tiền được giảm là: 1.297.500
Tổng tiền phải thanh toán: 7.352.500
Cảm ơn quý khách đã mua
"""
````

````python
chicken = int(input("Input the number of chicken: "))
hamburger = int(input("Input the number of hampurger: "))
pepsi = int(input("Input the number of pepsi: "))
total_bill = chicken*35000 + hamburger*50000 + pepsi*10000
print(f"Total bill is {total_bill}")

if total_bill >= 2000000:
    print(f"Total discount is {total_bill * 0.1}")
    print(f"Total bill that customer have to pay is {total_bill - (total_bill * 0.1)}")
elif total_bill >= 4000000:
    print(f"Total discount is {total_bill * 0.15}")
    print(f"Total bill that customer have top pay is {total_bill - (total_bill * 0.15)}")
print("Thank you for your purchase")
````

````python
chicken = int(input("Input the number of chicken: "))
hamburger = int(input("Input the number of hampurger: "))
pepsi = int(input("Input the number of pepsi: "))

total_bill = chicken*35000 + hamburger*50000 + pepsi*10000
discount = 0

print(f"Total bill is {total_bill}")

if total_bill >= 4000000:
    discount = 0.15 * total_bill
    total_bill -= discount 
    print(f"Total discount is {discount}")
elif total_bill >= 2000000:
    discount = 0.10 * total_bill
    total_bill -= discount
    print(f"Total bill is {discount}")

print(f"Total bill after discount is {total_bill}")
print("Thank you for your purchase")    
````

### Bài 11:

Nhập vào chiều cao (cm) và cân nặng (kg), tính số BMI và xét rồi xin kết quả theo dữ liệu sau:

- BMI < 16: Gầy cấp độ III
- 16 <= BMI < 17:  Gầy cấp độ II
- 17<= BMI < 18.5: Gầy cấp độ I
- 18.5 <= BMI < 25: Bình thường
- 25 <= BMI < 30: Thừa cân
- 30 <= BMI < 35 : Béo phì cấp độ I
- 35 <= BMI < 40: Béo phì cấp độ II
- BMI > 40: Béo phì cấp độ III

````python
heigh = int(input("Input heigh: "))
weigh = int(input("Input weigh: "))
bmi = weigh / (heigh/100) **2
print(f"BMI is {bmi}")

if bmi < 16:
    print("Gầy cấp độ III")
elif bmi < 17: 
    print("Gầy cấp độ II")
elif bmi < 18.5: 
    print("Gầy cấp độ I")
elif bmi < 25: 
    print("Bình thường")
elif bmi < 30: 
    print("Thừa cân")
elif bmi < 35: 
    print("Béo phì cấp độ I")
elif bmi < 40: 
    print("Béo phì cấp độ II")
else: 
    print("Béo phì cấp độ III")
````
