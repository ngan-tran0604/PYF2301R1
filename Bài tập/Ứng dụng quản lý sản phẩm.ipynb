```python
# Khai báo danh sách sản phẩm
prods_db = {}

# Hàm lấy thông tin sản phẩm bằng id sản phẩm
def get_product_info(prods_db, prod_id):
    if prod_id in prods_db:
        return prods_db[prod_id]
    else:
        return None

# Hàm cập nhật danh sách sản phẩm
def update_product_info(prods_db, prod_id, prod_name):
    prods_db[prod_id] = prod_name

# Hiển thị danh sách sản phẩm
print("Danh sách sản phẩm:")
print(prods_db)

# Vòng lặp quản lý sản phẩm
while True:
    # Hiển thị menu lựa chọn
    print("Menu:")
    print("1. Thêm sản phẩm mới")
    print("2. Cập nhật sản phẩm")
    print("3. Xoá sản phẩm")
    print("4. Thoát")

    # Nhập lựa chọn từ người dùng
    option = int(input("Nhập lựa chọn của bạn: "))

    # Xử lý lựa chọn từ người dùng
    if option == 1:
        # Thêm sản phẩm mới
        prod_id = int(input("Nhập ID sản phẩm: "))
        if get_product_info(prods_db, prod_id) is not None:
            print("Sản phẩm đã tồn tại với ID này.")
        else:
            prod_name = input("Nhập tên sản phẩm: ")
            update_product_info(prods_db, prod_id, prod_name)
            print("Thêm sản phẩm mới thành công.")

    elif option == 2:
        # Cập nhật sản phẩm
        prod_id = int(input("Nhập ID sản phẩm: "))
        prod_info = get_product_info(prods_db, prod_id)
        if prod_info is not None:
            prod_name = input("Nhập tên sản phẩm mới: ")
            update_product_info(prods_db, prod_id, prod_name)
            print("Cập nhật sản phẩm thành công.")
        else:
            print("Sản phẩm không tồn tại.")

    elif option == 3:
        # Xoá sản phẩm
        prod_id = int(input("Nhập ID sản phẩm: "))
        prod_info = get_product_info(prods_db, prod_id)
        if prod_info is not None:
            del prods_db[prod_id]
            print("Xoá sản phẩm thành công.")
        else:
            print("Sản phẩm không tồn tại.")

    elif option == 4:
        # Thoát
        break

    else:
     # Lựa chọn không hợp lệ
        print("Lựa chọn không hợp lệ.")

# Hiển thị danh sách sản phẩm
print("Danh sách sản phẩm:")
print(prods_db)
```
