```python
import turtle
import time

class Clock:
    def __init__(self):
        self.hour = 0
        self.minute = 0
        self.second = 0

    def draw_clock_face(self):
        # Vẽ mặt đồng hồ
        turtle.pensize(4)
        turtle.penup()
        turtle.goto(0, 0)
        turtle.pendown()
        turtle.circle(100)

        # Vẽ các vạch chia của đồng hồ
        turtle.penup()
        turtle.goto(0, 0)
        turtle.pensize(2)
        for i in range(12):
            turtle.penup()
            turtle.goto(0, 0)
            turtle.right(i * 30)
            turtle.forward(80)
            turtle.pendown()
            turtle.forward(10)

    def draw_hands(self):
        # Vẽ kim giờ
        turtle.penup()
        turtle.goto(0, 0)
        turtle.pensize(6)
        turtle.right(self.hour * 30 + self.minute / 2)
        turtle.pendown()
        turtle.forward(50)

        # Vẽ kim phút
        turtle.penup()
        turtle.goto(0, 0)
        turtle.pensize(4)
        turtle.right(self.minute * 6)
        turtle.pendown()
        turtle.forward(70)

        # Vẽ kim giây
        turtle.penup()
        turtle.goto(0, 0)
        turtle.pensize(2)
        turtle.right(self.second * 6)
        turtle.pendown()
        turtle.forward(90)

    def move_hands(self):
        while True:
            # Lấy thời gian thực
            current_time = time.localtime()
            self.hour = current_time.tm_hour % 12
            self.minute = current_time.tm_min
            self.second = current_time.tm_sec

            # Vẽ lại mặt đồng hồ và các kim
            turtle.clear()
            self.draw_clock_face()
            self.draw_hands()
            turtle.update()
            time.sleep(1)

# Khởi tạo đối tượng clock và chạy chương trình
clock = Clock()
clock.move_hands()

# Giữ màn hình turtle không đóng lại sau khi chạy xong chương trình
turtle.mainloop()
```
