```python
import tkinter as tk

expression = ""

def update_expression(value):
    global expression
    expression = expression + str(value)
    equation.set(expression)

def calculate():
    try:
        global expression
        total = str(eval(expression))
        equation.set(total)
        expression = ""
    except:
        equation.set("error")
        expression = ""

def clear():
    global expression
    expression = ""
    equation.set("")

root = tk.Tk()
root.title("Calculator")

root.configure(background="light blue")

root.geometry("250x200")

equation = tk.StringVar()
equation.set("")

expression_field = tk.Entry(root, textvariable=equation)
expression_field.grid(columnspan=4, ipadx=70)

button_1 = tk.Button(root, text="1", command=lambda: update_expression(1), bg="white", fg="black", height=2, width=7)
button_1.grid(row=1, column=0)

button_2 = tk.Button(root, text="2", command=lambda: update_expression(2), bg="white", fg="black", height=2, width=7)
button_2.grid(row=1, column=1)

button_3 = tk.Button(root, text="3", command=lambda: update_expression(3), bg="white", fg="black", height=2, width=7)
button_3.grid(row=1, column=2)

button_plus = tk.Button(root, text="+", command=lambda: update_expression("+"), bg="white", fg="black", height=2, width=7)
button_plus.grid(row=1, column=3)

button_4 = tk.Button(root, text="4", command=lambda: update_expression(4), bg="white", fg="black", height=2, width=7)
button_4.grid(row=2, column=0)

button_5 = tk.Button(root, text="5", command=lambda: update_expression(5), bg="white", fg="black", height=2, width=7)
button_5.grid(row=2, column=1)

button_6 = tk.Button(root, text="6", command=lambda: update_expression(6), bg="white", fg="black", height=2, width=7)
button_6.grid(row=2, column=2)

button_minus = tk.Button(root, text="-", command=lambda: update_expression("-"), bg="white", fg="black", height=2, width=7)
button_minus.grid(row=2, column=3)

button_7 = tk.Button(root, text="7", command=lambda: update_expression(7), bg="white", fg="black", height=2, width=7)
button_7.grid(row=3, column=0)

button_8 = tk.Button(root, text="8", command=lambda: update_expression(8), bg="white", fg="black", height=2, width=7)
button_8.grid(row=3, column=1)

button_9 = tk.Button(root, text="9", command=lambda: update_expression(9), bg="white", fg="black", height=2, width=7)
button_9.grid(row=3, column=2)

button_multiply = tk.Button(root, text="*", command=lambda: update_expression("*"), bg="white", fg="black", height=2, width=7)
button_multiply.grid(row=3, column=3)

button_clear = tk.Button(root, text="C", command=clear, bg="white", fg="black", height=2, width=7)
button_clear.grid(row=4, column=0)

button_0 = tk.Button(root, text="0", command=lambda: update_expression(0), bg="white", fg="black", height=2, width=7)
button_0.grid(row=4, column=1)

button_divide = tk.Button(root, text="/", command=lambda: update_expression("/"), bg="white", fg="black", height=2, width=7)
button_divide.grid(row=4, column=3)

button_decimal = tk.Button(root, text=".", command=lambda: update_expression("."), bg="white", fg="black", height=2, width=7)
button_decimal.grid(row=4, column=2)

button_equals = tk.Button(root, text="=", command=calculate, bg="white", fg="black", height=2, width=7)
button_equals.grid(row=5, column=2)

root.mainloop()
```
